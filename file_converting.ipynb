```python
import os
import subprocess

# Set the folder path containing .dav files
input_folder = "/Users/mouriya/Downloads/cashlifting/normal_cash_activities"  # 🔁 Change this to your actual folder path

# Create an output folder (optional)
output_folder = os.path.join(input_folder, "normal_converted_mp4")
os.makedirs(output_folder, exist_ok=True)

# Loop through all .dav files and convert them
for filename in os.listdir(input_folder):
    if filename.endswith(".dav"):
        input_path = os.path.join(input_folder, filename)
        output_filename = filename.replace(".dav", ".mp4")
        output_path = os.path.join(output_folder, output_filename)

        command = [
            "ffmpeg",
            "-i", input_path,
            "-c:v", "libx264",
            "-crf", "23",
            "-preset", "fast",
            output_path
        ]

        print(f"Converting {filename} → {output_filename}")
        subprocess.run(command)

print("✅ All .dav files converted to .mp4!")
```

```python
import os
import subprocess

# Input folder with long normal videos
input_folder = "/Users/mouriya/Downloads/cashlifting/normal_videos"

# Output folder for split 15-second videos
output_folder = os.path.join(input_folder, "split_15sec")
os.makedirs(output_folder, exist_ok=True)

# Loop through all .mp4 files
for filename in os.listdir(input_folder):
    if filename.endswith(".mp4"):
        input_path = os.path.join(input_folder, filename)

        # Output name pattern: filename_part_001.mp4, etc.
        base_name = os.path.splitext(filename)[0]
        output_pattern = os.path.join(output_folder, f"{base_name}_part_%03d.mp4")

        # FFmpeg command to split into 15s chunks without re-encoding
        command = [
    "ffmpeg",
    "-i", input_path,
    "-c:v", "libx264",
    "-crf", "23",
    "-preset", "fast",
    "-force_key_frames", "expr:gte(t,n_forced*15)",
    "-f", "segment",
    "-segment_time", "15",
    "-c:a", "aac",
    "-strict", "experimental",
    output_pattern
]


        print(f"⏳ Splitting: {filename}")
        subprocess.run(command)

print("✅ All videos split into 15-second chunks!")
```

```python
import os
import subprocess
import re

input_folder = "/Users/mouriya/Downloads/cashlifting/normal_videos"
output_folder = os.path.join(input_folder, "split_clean_15sec")
os.makedirs(output_folder, exist_ok=True)

def get_black_end_time(video_path):
    try:
        result = subprocess.run(
            ["ffmpeg", "-i", video_path, "-vf", "blackdetect=d=0.5:pic_th=0.98", "-an", "-f", "null", "-"],
            stderr=subprocess.PIPE,
            stdout=subprocess.PIPE,
            text=True
        )
        output = result.stderr
        matches = re.findall(r'black_end:(\d+\.\d+)', output)
        if matches:
            return float(matches[-1])  # Use last black_end time
    except Exception as e:
        print(f"Error detecting black frames: {e}")
    return 0.0

for filename in os.listdir(input_folder):
    if filename.endswith(".mp4"):
        input_path = os.path.join(input_folder, filename)
        base_name = os.path.splitext(filename)[0]
        output_pattern = os.path.join(output_folder, f"{base_name}_part_%03d.mp4")

        # Step 1: Detect how long black screen lasts
        black_end = get_black_end_time(input_path)
        print(f"{filename} ➜ Skipping first {black_end:.1f} seconds")

        # Step 2: Skip black frames and split from there
        command = [
            "ffmpeg",
            "-ss", str(black_end),
            "-i", input_path,
            "-c:v", "libx264",
            "-crf", "23",
            "-preset", "fast",
            "-force_key_frames", "expr:gte(t,n_forced*15)",
            "-f", "segment",
            "-segment_time", "15",
            "-c:a", "aac",
            output_pattern
        ]

        subprocess.run(command)

print("✅ Done splitting all videos after skipping black screens!")
```

```python
import os
import subprocess
import re

input_folder = "/Users/mouriya/Downloads/cashlifting/normal_videos"
output_folder = os.path.join(input_folder, "clean_15s_clips")
os.makedirs(output_folder, exist_ok=True)

def get_black_segments(video_path):
    result = subprocess.run(
        ["ffmpeg", "-i", video_path, "-vf", "blackdetect=d=1:pic_th=0.98", "-an", "-f", "null", "-"],
        stderr=subprocess.PIPE, text=True
    )
    black_segments = []
    for line in result.stderr.split('\n'):
        match = re.search(r'black_start:(\d+(\.\d+)?) black_end:(\d+(\.\d+)?)', line)
        if match:
            black_segments.append((float(match.group(1)), float(match.group(3))))
    return black_segments

def get_video_duration(video_path):
    result = subprocess.run(
        ["ffprobe", "-v", "error", "-show_entries",
         "format=duration", "-of",
         "default=noprint_wrappers=1:nokey=1", video_path],
        stdout=subprocess.PIPE, stderr=subprocess.STDOUT
    )
    return float(result.stdout.decode().strip())

def get_visible_segments(black_segments, total_duration):
    visible_segments = []
    start = 0
    for black_start, black_end in black_segments:
        if black_start - start >= 15:
            visible_segments.append((start, black_start))
        start = black_end
    if total_duration - start >= 15:
        visible_segments.append((start, total_duration))
    return visible_segments

def extract_15s_chunks(video_path, visible_segments, base_name):
    count = 0
    for start, end in visible_segments:
        segment_duration = end - start
        num_clips = int(segment_duration // 15)
        for i in range(num_clips):
            clip_start = start + i * 15
            output_path = os.path.join(output_folder, f"{base_name}_clip_{count:03d}.mp4")
            command = [
                "ffmpeg", "-ss", str(clip_start), "-i", video_path,
                "-t", "15", "-c:v", "libx264", "-crf", "23", "-preset", "fast",
                output_path
            ]
            subprocess.run(command)
            count += 1

for filename in os.listdir(input_folder):
    if filename.endswith(".mp4"):
        input_path = os.path.join(input_folder, filename)
        base_name = os.path.splitext(filename)[0]

        print(f"⏳ Processing {filename}")
        total_duration = get_video_duration(input_path)
        black_segments = get_black_segments(input_path)
        visible_segments = get_visible_segments(black_segments, total_duration)
        extract_15s_chunks(input_path, visible_segments, base_name)

print("✅ Finished extracting clean 15-second clips from all videos.")
```
